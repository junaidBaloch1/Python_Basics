### 7 Python Decorators


Powerful tools of python. They are used to modify the behaviours of functions. Provide the flexibility to wrap another function to expand the working of wrapped function without permanently modifying it.
Simply we expand the working of already defined function by writing more code to it without affecting the working of its previous version.
#### In Decorators, functions are passed as an argument into another function and then called inside the wrapper function.
1. The function can be referenced and passed to a variable and returned from other functions as well.
2. The functions can be declared inside another function and passed as an argument to another function.

#### 7.2 Higher order functions


A function that accept another function as an argument is called higher order function.

```python
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def operation(func,a,b):
    temp = func(a,b)
    return temp

print("add up",operation(add,5,12))
print("subtract them",operation(sub,5,12))



def greet():
    print("Hi junaid")
    
    def subgreet():
        print("hi!")
    return subgreet

obj = greet()
obj()
    
```

#### 7.3 Decorating functions with parameters

```python
def add(x,y):  
    return x+y  
def test(func):  
    def inner(x,y):  
        if(x<y):  
            x,y = y,x 
            return func(x,y)  
    return inner  

result = test(add)  
result(2,4) 
```

#### 7.4 syntactic decorator

```python
def test(func):  
    def inner(x,y):  
        if(x<y):  
            x,y = y,x 
            return func(x,y)  
    return inner 
@test

def add(x,y):  
    return x+y 

add(4,13)
```

#### 7.5 Reusing Decorators

```python
def times(func):
    def wrap():
        func()
        func()
    return wrap
@times

def greet():
    print("it prints result two time")
    
greet()
```

#### 7.6 Python decorators with argument

```python
# Modify the above code,| args* kwargs**

def times(func):
    def wrap(*args,**kwargs):
        func(*args,**kwargs)
        func(*args,**kwargs)
    return wrap




@times

def display(name,gpa):
    print(f"hello{name}{gpa}")
display(' junaid ',2.97)
```

### 8 Python Lambda function
